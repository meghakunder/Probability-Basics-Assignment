**Question 3: A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability**
**that the next ball is also red (without replacement)?**

**Ans:**
Since one red ball is already taken out:

Remaining red balls =3-1=2

Remaining total balls =5-1=4

P(next ball is red)=Remaining Red balls/Total Remaining Balls
                   =2/4=1/2
                   =0.5

So, the probability that the second ball is also red is 50%.

**Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation**
**of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?**
**Why?**

**Ans:**
The school population is not evenly distributed:

Boys = 60%

Girls = 40%

If  Simple Random Sampling is used , it may accidentally select more boys and fewer girls, because boys are the majority. This would not guarantee equal representation.

With Stratified Sampling:

First divide the population into two groups (strata): Boys and Girls.

Then select equal numbers from each group.

This guarantees fair and equal representation of both genders in the sample.

Stratified sampling should be used because the population is divided into groups (boys and girls), and this method ensures equal representation from each group, which simple random sampling does not guarantee.

**Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the probability that all are non-defective?**
**Ans:**
Defective rate = 2% = 0.02

So, Non-defective rate = 1 - 0.02 = 0.98

Number of bulbs chosen = 5

P(All non-defective)=(0.98)^5
                     =**0.9039 or 90.39%**

**Question 10: Differentiate between discrete and continuous random variables with examples**

**Ans:**

**1. Discrete Random Variable**
A discrete random variable takes only separate, countable values.
These values are usually whole numbers.

* Can be counted

* Has gaps between values

* Mostly comes from counting

**Examples:**

Number of students in a class

Number of heads in 5 coin tosses

Number of defective bulbs in a box

Number of cars in a parking lot

**2. Continuous Random Variable**

A continuous random variable can take any value within a given range.
These values are measured, not counted.


* Cannot be counted exactly

* No gaps between values

* Mostly comes from measurement

**Examples:**

Height of students

Weight of a person

Time taken to complete a test

Temperature of a room
